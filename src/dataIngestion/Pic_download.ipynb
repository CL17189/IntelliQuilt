```python
import pandas as pd
import requests
import os
import time
import numpy as np
from PIL import Image
from io import BytesIO
import ast
import numpy as np
```

```python
df=pd.read_csv('cleanedData.csv')
df.head()
```

```python
df['image_link'][3979]
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36'
}

# 使用 Session
session = requests.Session()
session.headers.update(headers)
```

```python
folder = 'preface_pic'
os.makedirs(folder, exist_ok=True)

# 定义下载的区域
start_row = 1  # 开始的行号（包含）
end_row = len(df['image_link'])   # 结束的行号（包含）

# 遍历 DataFrame 下载图片
for index, row in df.iterrows():
    if start_row <= index+1 <= end_row:  # index+1 是因为 DataFrame 的索引默认从0开始
        image_url = row['image_link']
        filename = f"{folder}/{row['num']}.jpg"

        # 检查文件是否已存在
        if not os.path.exists(filename):

            try:
                res=session.get(image_url)
                if res.status_code == 200:
                    image = Image.open(BytesIO(res.content))

                    # 保存图片到本地
                    image.save(filename)
                    print(f"Image {row['num']} downloaded successfully.")
                else:
                    print(f"Failed to download image {row['num']}.")
                time.sleep(1)

            except Exception as e:
                print(f"An error occurred while downloading image {row['num']}: {e}")
                time.sleep(1)

        else:
            pass
            #print(f"Image {row['num']} already exists. Skipping download.")
        
```

```python
folder = 'detailed_pic'
os.makedirs(folder, exist_ok=True)

# 定义下载的区域
start_row = 3259 # 开始的行号（包含）
end_row = len(df_t['src_list'])   # 结束的行号（包含）

# 遍历 DataFrame 下载图片
for index, row in df_t.iterrows():
    if start_row <= index+1 <= end_row and not pd.isna(row['src_list']) :  # index+1 是因为 DataFrame 的索引默认从0开始
        url_list = row['src_list']
        
        for num,link in enumerate(ast.literal_eval(url_list)):

            filename = f"{folder}/{row['num']}_{num}.jpg"

            # 检查文件是否已存在
            if not os.path.exists(filename):

                try:
                    res=session.get(link)
                    if res.status_code == 200:
                        image = Image.open(BytesIO(res.content))

                        # 保存图片到本地
                        image.save(filename)
                        print(f"Image {row['num']}_{num} downloaded successfully.")
                    else:
                        print(f"Failed to download image {row['num']}_{num}.")

                except Exception as e:
                    print(f"An error occurred while downloading image {row['num']}_{num}: {e}")
                    #time.sleep(1)

            else:
                print(f"Image {row['num']}_{num} already exists. Skipping download.")
            if num% 50==0:time.sleep(3)
```
