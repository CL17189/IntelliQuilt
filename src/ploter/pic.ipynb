```python
import matplotlib.pyplot as plt
import numpy as np
# Replot the graph with requested modifications
plt.rcParams['font.sans-serif'] = ['SimHei'] # sans-serif
plt.rcParams['axes.unicode_minus'] = False   # 
# Data
years = [2018, 2019, 2020, 2021, 2022, 2023]
sales = [7.0, 8.5, 9.8, 10.8, 11.5, 13.0]
growth_rates = [25.4, 19.5, 14.8, 12.0, 6.2, 8.4]

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plotting the bar chart
color = 'tab:orange'
ax1.set_xlabel('')
ax1.set_ylabel('', color=color)
bars = ax1.bar(years, sales, color=color, alpha=0.6, label='')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 15)

# Add data labels on bars
for bar in bars:
    height = bar.get_height()
    ax1.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Create a second y-axis to plot the line chart
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('%', color=color)
line = ax2.plot(years, growth_rates, color=color, marker='o', label='')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, 30)

# Add data labels on line points
for i, txt in enumerate(growth_rates):
    ax2.annotate(f'{txt}', (years[i], growth_rates[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Add a combined legend
bars_labels = ['']
line_labels = ['']
labels = bars_labels + line_labels
handles = [bars, line[0]]

ax1.legend(handles, labels, loc='upper left')

# Add title and show plot
plt.title('2018-2023')
fig.tight_layout()  
plt.show()
```
