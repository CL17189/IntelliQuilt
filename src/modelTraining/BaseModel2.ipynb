```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Mock class to simulate the train_objects that user might have
class TrainObject:
    def __init__(self, max_accuracy, model_complexity, convergence_speed, training_speed, generalization):
        self.model_info = {
            'max_accuracy': max_accuracy,
            'model_complexity': model_complexity,
            'convergence_speed': convergence_speed,
            'training_speed': training_speed,
            'generalization': generalization
        }

# Create a list of TrainObject instances with dummy data
train_objects = [
    TrainObject(0.8, 0.6, 0.5, 0.7, 0.9),
    TrainObject(0.75, 0.7, 0.6, 0.8, 0.85)
]
model_names = ['Model 1', 'Model 2']

# Using seaborn style
sns.set(style="whitegrid")

# Setting the radar chart labels and angles
labels = ['Accuracy', 'Model Complexity', 'Convergence Speed', 'Training Speed', 'Generalization']
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the plot

# Create the figure for plotting
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each model's performance in the radar chart
for train_obj, model_name in zip(train_objects, model_names):
    stats = train_obj.model_info
    values = [stats['max_accuracy'], 
              stats['model_complexity'], 
              stats['convergence_speed'], 
              stats['training_speed'], 
              stats['generalization']]
    values += values[:1]  # close the values list
    ax.plot(angles, values, label=model_name, linewidth=2)
    ax.fill(angles, values, alpha=0.25)

# Set the radar chart properties
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Set the labels for the radar chart
plt.xticks(angles[:-1], labels, color='grey', size=12)

# Add a title and a legend to the chart
plt.title('Model Comparison', size=20, color='black', y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Display the plot
plt.show()
```
